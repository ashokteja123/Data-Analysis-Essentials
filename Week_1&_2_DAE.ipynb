Expand Dims & Squeeze

In deep learning, models don't process one image at a time — they process batches. An image is shape (224,224,3). A batch of 32 images is (32,224,224,3). That extra 32 at the front is added by expand_dims. You will do this every single time you feed an image to a neural network.

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Original:", arr.shape)        # (2,3)
```

```python
# Add axis at position 0 — add a "batch" dimension
e0 = np.expand_dims(arr, axis=0)
print("Axis 0:", e0.shape)           # (1,2,3)
```

```python
# Add axis at position 1
e1 = np.expand_dims(arr, axis=1)
print("Axis 1:", e1.shape)           # (2,1,3)
```

```python
# Add axis at position 2
e2 = np.expand_dims(arr, axis=2)
print("Axis 2:", e2.shape)           # (2,3,1)
```

axis=0 adds a new outermost layer → think "batch of 1"
axis=1 adds between rows and columns → think "each row is its own group"
axis=2 adds innermost → think "each element becomes a list of 1"

Squeeze

```python
# Squeeze removes dimensions of size 1
a = np.zeros((1, 3, 1, 4))
print("Before squeeze:", a.shape)     # (1,3,1,4)
```

```python
s = np.squeeze(a)
print("After squeeze:", s.shape)      # (3,4) — both size-1 dims removed
```

```python
# Squeeze specific axis only
s1 = np.squeeze(a, axis=0)
print("Squeeze axis 0:", s1.shape)    # (3,1,4)
```

```python
s2 = np.squeeze(a, axis=2)
print("Squeeze axis 2:", s2.shape)    # (1,3,4)
```

"squeeze() only removes dimensions whose size is 1 — it never removes a dimension that actually has data. So squeeze on a (3,4) array does nothing — there's nothing to remove."

```python
b = np.zeros((3, 4))
print(np.squeeze(b).shape)   # still (3,4) — nothing removed
```

```python
marks = np.array([78, 45, 92, 63, 88, 51])

print(np.sort(marks))              # ascending by default
print(np.sort(marks)[::-1])        # descending — reverse the sorted array
print(np.argsort(marks))           # indices that would sort the array

# Real use: who is the topper?
names = ["Ravi", "Priya", "Suresh", "Anita", "Kiran", "Divya"]
rank_order = np.argsort(marks)[::-1]
print("Rank 1:", names[rank_order[0]], marks[rank_order[0]])
```

Why ones instead of zeros?

When you multiply any number by 1, it stays the same.
When you add 0 to any number, it stays the same.

These are called identity elements.
0 is the identity element for addition.
1 is the identity element for multiplication.

In mathematics, starting with ones is useful when:
- You are tracking multipliers (no change = 1, double = 2, half = 0.5)
- You are creating a baseline where everything starts equally weighted
- You are building a mask where everything is initially included

Example:
You give all 40 students equal weight = 1 in a voting system.
Later you might change some weights based on performance.
Starting point = all ones.text*

```python
import numpy as np

# 12 months of bank interest calculation
# Multiplier starts at 1 (no interest yet applied)
monthly_multiplier = np.ones(12)
print("Base multipliers:", monthly_multiplier)

# Apply 8% annual interest = 0.08/12 per month added
monthly_rate = 0.08 / 12
monthly_multiplier = 1 + monthly_rate
print("Monthly interest multiplier:", monthly_multiplier)

# Starting deposit of Rs 10000
deposit = 10000

# Calculate balance each month using ones as starting structure
balances = np.ones(12) * deposit
for i in range(1, 12):
    balances[i] = balances[i-1] * (1 + monthly_rate)

print("Monthly balances:")
for i, bal in enumerate(balances):
    print(f"Month {i+1}: Rs {bal:.2f}")
```

Example 5 — GST Rate Baseline

```python
import numpy as np

# 10 products — base GST multiplier
# Before GST: multiplier = 1 (price unchanged)
# After applying 18% GST: multiplier = 1.18
base_prices = np.array([500, 1200, 350, 2500, 180, 890, 3200, 750, 420, 1100])
print("Base prices:", base_prices)

# All products start with multiplier of 1
gst_multiplier = np.ones(10)

# Some products have 18% GST, some have 5%, some are exempt
gst_18_products = [0, 1, 4, 7]    # indices of products with 18% GST
gst_5_products = [2, 3, 5, 8]     # indices of products with 5% GST
# remaining [6, 9] are exempt — multiplier stays 1

gst_multiplier[gst_18_products] = 1.18
gst_multiplier[gst_5_products] = 1.05

print("GST multipliers:", gst_multiplier)
final_prices = base_prices * gst_multiplier
print("Final prices after GST:", final_prices.round(2))
```

Where do random numbers come from in computers?

Computers are deterministic machines.
They always do exactly what you tell them.
They cannot actually be random.

Instead they use a formula called a pseudo-random number generator.
You give it a starting number called a SEED.
The formula produces a sequence that LOOKS random but is predictable.

Why does this matter?
If you run the same code with the same seed, you get the same random numbers.
This is useful for:
- Research: others can reproduce your exact results
- Testing: your program gives same output every time for verification
- Debugging: you can recreate the exact situation that caused an error

Three types of random in NumPy:

1. np.random.rand()    → decimals between 0.0 and 1.0
2. np.random.randint() → whole numbers in a range you specify
3. np.random.randn()   → follows normal distribution (bell curve)
                         most values near 0, fewer far from 0

For most practical work with marks, prices, ages: use randint()
For probability and statistics work: use rand() and randn()

*
           *  *  *
        *  *  *  *  *
     *  *  *  *  *  *  *
  *  *  *  *  *  *  *  *  *
 -3  -2  -1   0   1   2   3

Most values fall in the middle.
Very few values are far from zero.
This is how most natural measurements behave.
(Height, weight, exam marks of a large class.)

Example 2 — Flatten vs Ravel (the Difference)

```python
import numpy as np

original = np.array([[10, 20, 30],
                     [40, 50, 60]])

flat_copy = original.flatten()
ravel_view = original.ravel()

print("Original:", original.flatten())
print("Flattened copy:", flat_copy)
print("Raveled view:", ravel_view)

# Change the raveled view
ravel_view[0] = 999
print("\nAfter changing ravel_view[0] to 999:")
print("ravel_view:", ravel_view)
print("original (ALSO CHANGED because ravel shares memory):", original)

# Now try with flatten — original should NOT change
original2 = np.array([[10, 20, 30], [40, 50, 60]])
flat_copy2 = original2.flatten()
flat_copy2[0] = 999
print("\nAfter changing flat_copy2[0] to 999:")
print("flat_copy2:", flat_copy2)
print("original2 (UNCHANGED because flatten makes a copy):", original2)
```

Program f — Transpose of a NumPy Array

Example 1 — Student Marks Transpose

How can you transpose student marks data from a student-subject view to a subject-student view and calculate subject averages?

```python
import numpy as np

# Original: rows = students, columns = subjects
marks = np.array([[85, 92, 78],
                  [70, 65, 90],
                  [95, 88, 76],
                  [60, 72, 68]])

print("Original (students × subjects):")
print(marks)
print("Shape:", marks.shape, "→ 4 students, 3 subjects")

marks_T = marks.T
print("\nTransposed (subjects × students):")
print(marks_T)
print("Shape:", marks_T.shape, "→ 3 subjects, 4 students")

print("\nNow each ROW is a subject:")
subject_names = ["Maths", "Physics", "Chemistry"]
for i, subject in enumerate(subject_names):
    print(f"  {subject} scores: {marks_T[i]} → average: {marks_T[i].mean():.1f}")
```

Example 2 — Monthly Revenue Transpose

How can you transpose monthly revenue data from a month-product view to a product-month view to analyze product performance?

```python
import numpy as np

# Original: rows = months (12), columns = products (4)
np.random.seed(18)
revenue = np.random.randint(50000, 200000, size=(12, 4))

print("Original shape (months × products):", revenue.shape)
print("First 3 months:")
print(revenue[:3])

revenue_T = revenue.T
print("\nTransposed shape (products × months):", revenue_T.shape)
print("Product 1 revenue across all 12 months:")
print(revenue_T[0])
print("Product 1 total annual revenue:", revenue_T[0].sum())
print("Product 1 best month revenue:", revenue_T[0].max())
```

Example 3 — Transpose for Column Operations

How can you perform column-wise operations on a 2D array, such as calculating subject averages, using both the `axis` parameter and array transposition?

```python
import numpy as np

data = np.array([[85, 92, 78, 88],
                 [70, 65, 90, 82],
                 [95, 88, 76, 91]])

print("Original data:")
print(data)

# Subject-wise statistics (operating on columns)
# Method 1: using axis
print("\nSubject averages (axis=0):", data.mean(axis=0))

# Method 2: using transpose — sometimes more intuitive
print("Subject averages (via transpose):", data.T.mean(axis=1))

# Both give same result — shows two approaches to same problem
print("Are they equal?", np.allclose(data.mean(axis=0), data.T.mean(axis=1)))
```

Example 4 — Transpose to Align for Multiplication

How can you calculate the weighted average of student marks across multiple subjects using NumPy's broadcasting capabilities?

```python
import numpy as np

# student marks: 3 students × 5 subjects
marks = np.array([[85, 92, 78, 88, 76],
                  [70, 65, 90, 82, 79],
                  [95, 88, 76, 91, 84]])

# subject weights (5 weights for 5 subjects)
weights = np.array([0.25, 0.20, 0.20, 0.20, 0.15])

print("Marks shape:", marks.shape)
print("Weights shape:", weights.shape)

# Weighted average for each student
# marks is (3,5), weights is (5,) → broadcasting works
weighted_scores = marks * weights   # broadcasts weights across rows
weighted_average = weighted_scores.sum(axis=1)

print("\nWeighted average per student:")
for i, score in enumerate(weighted_average):
    print(f"  Student {i+1}: {score:.1f}")
```

Example 5 — Double Transpose Returns Original

What happens when you transpose a NumPy array twice, and how can you verify if it returns the original array?

```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original:")
print(arr)
print("Shape:", arr.shape)

transposed = arr.T
print("\nTransposed:")
print(transposed)
print("Shape:", transposed.shape)

double_transposed = arr.T.T
print("\nDouble transposed (should equal original):")
print(double_transposed)
print("Shape:", double_transposed.shape)
print("Equal to original?", np.array_equal(arr, double_transposed))
```
